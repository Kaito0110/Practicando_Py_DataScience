### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion_tienda_1 = tienda["Precio"].sum()
facturacion_tienda_2 = tienda2["Precio"].sum()
facturacion_tienda_3 = tienda3["Precio"].sum()
facturacion_tienda_4 = tienda4["Precio"].sum()

print("Facturación Tienda 1:", facturacion_tienda_1)
print("Facturación Tienda 2:", facturacion_tienda_2)
print("Facturación Tienda 3:", facturacion_tienda_3)
print("Facturación Tienda 4:", facturacion_tienda_4)
```

```python
facturacion = {
    "Tienda 1": facturacion_tienda_1,
    "Tienda 2": facturacion_tienda_2,
    "Tienda 3": facturacion_tienda_3,
    "Tienda 4": facturacion_tienda_4
}

facturacion
```

# 2. Ventas por categoría

```python
categorias_tienda_1 = tienda["Categoría del Producto"].value_counts()
categorias_tienda_2 = tienda2["Categoría del Producto"].value_counts()
categorias_tienda_3 = tienda3["Categoría del Producto"].value_counts()
categorias_tienda_4 = tienda4["Categoría del Producto"].value_counts()

print("Tienda 1")
print(categorias_tienda_1)

print("\nTienda 2")
print(categorias_tienda_2)

print("\nTienda 3")
print(categorias_tienda_3)

print("\nTienda 4")
print(categorias_tienda_4)
```

```python
print("Categoría más vendida en Tienda 1:", categorias_tienda_1.idxmax())
print("Categoría más vendida en Tienda 2:", categorias_tienda_2.idxmax())
print("Categoría más vendida en Tienda 3:", categorias_tienda_3.idxmax())
print("Categoría más vendida en Tienda 4:", categorias_tienda_4.idxmax())
```

# 3. Calificación promedio de la tienda


```python
promedio_tienda_1 = tienda["Calificación"].mean()
promedio_tienda_2 = tienda2["Calificación"].mean()
promedio_tienda_3 = tienda3["Calificación"].mean()
promedio_tienda_4 = tienda4["Calificación"].mean()

print("Promedio Tienda 1:", promedio_tienda_1)
print("Promedio Tienda 2:", promedio_tienda_2)
print("Promedio Tienda 3:", promedio_tienda_3)
print("Promedio Tienda 4:", promedio_tienda_4)
```

```python
print("Calificación promedio de la Tienda 1:", round(promedio_tienda_1, 2))
print("Calificación promedio de la Tienda 2:", round(promedio_tienda_2, 2))
print("Calificación promedio de la Tienda 3:", round(promedio_tienda_3, 2))
print("Calificación promedio de la Tienda 4:", round(promedio_tienda_4, 2))
```

# 4. Productos más y menos vendidos

```python
productos_tienda_1 = tienda["Producto"].value_counts()
productos_tienda_2 = tienda2["Producto"].value_counts()
productos_tienda_3 = tienda3["Producto"].value_counts()
productos_tienda_4 = tienda4["Producto"].value_counts()

print("Producto más vendido Tienda 1:", productos_tienda_1.idxmax())
print("Producto más vendido Tienda 2:", productos_tienda_2.idxmax())
print("Producto más vendido Tienda 3:", productos_tienda_3.idxmax())
print("Producto más vendido Tienda 4:", productos_tienda_4.idxmax())
```

```python
print("Producto menos vendido Tienda 1:", productos_tienda_1.idxmin())
print("Producto menos vendido Tienda 2:", productos_tienda_2.idxmin())
print("Producto menos vendido Tienda 3:", productos_tienda_3.idxmin())
print("Producto menos vendido Tienda 4:", productos_tienda_4.idxmin())
```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda_1 = tienda["Costo de envío"].mean()
envio_promedio_tienda_2 = tienda2["Costo de envío"].mean()
envio_promedio_tienda_3 = tienda3["Costo de envío"].mean()
envio_promedio_tienda_4 = tienda4["Costo de envío"].mean()

print("Envío promedio Tienda 1:", envio_promedio_tienda_1)
print("Envío promedio Tienda 2:", envio_promedio_tienda_2)
print("Envío promedio Tienda 3:", envio_promedio_tienda_3)
print("Envío promedio Tienda 4:", envio_promedio_tienda_4)
```

```python
print("Envío promedio de la Tienda 1:", round(envio_promedio_tienda_1, 2))
print("Envío promedio de la Tienda 2:", round(envio_promedio_tienda_2, 2))
print("Envío promedio de la Tienda 3:", round(envio_promedio_tienda_3, 2))
print("Envío promedio de la Tienda 4:", round(envio_promedio_tienda_4, 2))
```
